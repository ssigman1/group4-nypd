```python
# query NYC 311 for driver misconduct
# value list
complaint_types_dm = ('Blocked Bike Lane', 'Blocked Crosswalk', 'Blocked Hydrant', 'Blocked Sidewalk', 'Chronic Speeding', 'Chronic Stoplight Violation','Detached Trailer', 'Double Parked Blocking Traffic',
       'Double Parked Blocking Vehicle', 'Drag Racing',
       'Overnight Commercial Storage', 'Parking Permit Improper Use',
       'Posted Parking Sign Violation', 'Truck Route Violation',
       'Unauthorized Bus Layover', 'Abandoned Vehicle With License Plate' )
query = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv?$query=" +  requests.utils.quote(f"SELECT * WHERE descriptor IN {complaint_types_dm}  LIMIT 3000000")
dm  = pd.read_csv(query)
```

#Creating Column "Duration"
* a column based on two other columns

```python
dm['created_date'] = pd.to_datetime(dm['created_date'])
dm['closed_date'] = pd.to_datetime(dm['closed_date'])
dm['duration'] = dm['closed_date'] - dm['created_date']
```

#Filter
Response times under 5 minutes
* pd.Timedelta()

    (one method we didn’t learn in class by reviewing documentation
from a Python library)
* len() and f string statement

```python
dm_filtered = dm[dm['duration'] <= pd.Timedelta(minutes=5)]
dm_filtered.sample(5)
```

```python
length_address = f"There are {len(dm_filtered['incident_address'].unique())} unique addresses in the filtered data frame."
print(length_address)
```

#Loop and Conditional Statement

```python
for index, row in dm_filtered.iterrows():
    duration = pd.to_timedelta(row['duration'])
    if duration.seconds < 60:
        duration = str(duration.seconds) + " seconds"
    else:
        minutes = duration.seconds // 60
        seconds = duration.seconds % 60
        duration = str(minutes) + " minutes and " + str(seconds) + " seconds"
    dm_filtered.at[index, 'duration'] = duration
```

```python
def get_day_of_week(date_str):
    date_obj = pd.to_datetime(date_str)
    return date_obj.day_name()


dm_filtered.loc[:,'day_of_week'] = dm_filtered['created_date'].apply(get_day_of_week)


day_counts = dm_filtered.groupby('day_of_week').size().reset_index(name='day_count')
```

# Spacial Join
* create 2 columns ('year' and 'day')

```python
dm_geo = gpd.GeoDataFrame(dm, geometry = gpd.points_from_xy(dm['longitude'], dm['latitude']), crs="EPSG:4326")
dm_geo_p = dm_geo.sjoin(police, how ="left", predicate="within")
dm_geo_p['year'] = dm_geo_p['created_date'].dt.year
dm_geo_p['day'] = dm_geo_p['created_date'].dt.day
```

```python
dm_geo_p.head(10)
```

**Tidy Data (dm_geo_p)**
* each column represents one variable (ex. incident_address, created_date, location_type, etc.)
* each row represents one observation (each complaint)

# Inner Join

```python
total_per_year = dm_geo_p.groupby(['year'])['count_row'].sum().reset_index()
```

```python
average_per_year = dm_geo_p.groupby(['year', 'day'])['count_row'].sum().reset_index().groupby('year')['count_row'].mean().reset_index()
```

```python
inner_year = pd.merge(total_per_year, average_per_year, how ='inner', on='year')
```

```python
inner_year.sample(3)
```

**Tidy Data (inner_year)**
* each variable forms a column (x and y)
* each observation forms a row (year)

# Melt & Pivot

```python
des_com_melt = pd.melt(dm_geo_p, id_vars=['unique_key'], value_vars=['complaint_type', 'descriptor'])
```

```python
des_com_melt.sort_values(by=['unique_key', 'variable']).head(6)
```

**Tidy Data (des_com_melt)**
* each variable does NOT form a column because 'complaint_type' and 'descriptor' both exist in the 'variable' column
* each observation forms TWO rows, one for 'complaint_type' and one for 'descriptor'

```python
des_com_pivot = des_com_melt.pivot(index='unique_key', columns='variable', values='value')
```

```python
des_com_pivot.sample(3)
```

**Tidy Data (des_com_pivot)**
